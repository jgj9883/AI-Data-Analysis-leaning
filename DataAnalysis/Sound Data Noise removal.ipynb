```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

y, sr = librosa.load('data/sound/0001_sound_data.wav')
ft = librosa.stft(y)

db = librosa.amplitude_to_db(np.abs(ft), ref=np.max)
librosa.display.specshow(db, x_axis='time', y_axis='linear')
plt.show()
```

```python
bg = np.c_[ft[:,500:1100], ft[:,1300:1800], ft[:,1950:2250],
           ft[:,2250:2500], ft[:,2500:]]
fg = np.c_[ft[:,0:5000], ft[:,1250:1400], ft[:,1850:2150],
           ft[:,2250:2500], ft[:,2500:]]

y = librosa.istft(fg)
librosa.output.write_wav('fan-fg.wav', y, sr)
```
