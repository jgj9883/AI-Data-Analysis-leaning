```python
import pandas as pd
import folium as g
```

```python
gps_df = pd.read_csv('./data/gps/butchery_gps.csv',engine='python',encoding='CP949')
```

```python
gps_df
```

```python
g_map = g.Map(location=[37.891107, 127.768548])
```

```python
g_map
```

```python
marker = g.Marker([37.891107, 127.768548], popup='테스트 농장', icon=g.Icon(color='blue'))
marker.add_to(g_map)
g_map
```

```python
for name, lat, lng in zip(gps_df['name'], gps_df['latitude'], gps_df['longitude']):
    g.Marker([lat, lng], popup=name, icon=g.Icon(color='blue')).add_to(g_map)
```
