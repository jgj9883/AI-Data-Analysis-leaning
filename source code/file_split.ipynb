### File Split

```python
import os 
import glob
import shutil

month = int(input("몇 월 데이터를 나누시겠습니까? : "))

if 0 < month < 10:
    month = '0' + str(month)

elif 10 <= month <= 12 : 
    month = str(month)
else :
    print("No data")
    
    
for i in range(1, 32) :
    number = 0
    if i < 10 :
        number = str(0) + str(i)
        print(f'2021{month}0{i}')
    else :
        number = str(i)
        print(f'2021{month}{i}')
        
    data_dir = f'C:/Users/USER/Desktop/데이터/체리부로 이미지/체리부로 {month}월 데이터/2021{month}'+str(number)
    # print(data_dir)
    result_dir = f'C:/Users/USER/Desktop/test_data/{month}월/'
    FileList = glob.glob(data_dir + '/*.jpg')
    img_namess = FileList
    
    if not os.path.exists(result_dir) :
        os.mkdir(result_dir)
    
    count = 0
    result_list = []
    temp_module_index = ''
    for idx, d in enumerate(img_namess):
        FA_index = d.rfind('FA')
        h01_index = d.rfind('H01')
        h02_index = d.rfind('H02')
        date_index = d.rfind('2021')
        module_index = d[FA_index+16 : FA_index + 20]
        
        if count >= 1  :
            count = 0
            continue
        
        if module_index == '0001' or module_index == '0002' or module_index =='0007'or module_index =='0010' or module_index =='0013' or module_index =='0015' :
            
            if (module_index == temp_module_index):
                continue
                
            if h01_index > 0 :

                hour_index = d[date_index+8:date_index+10]
                if hour_index == '13':
                    shutil.copyfile(d, result_dir+d[FA_index:])
                    result_list.append(d[FA_index:])
                    count+=1
                    temp_module_index = module_index
                    print(count, d[FA_index:])

            elif h02_index > 0 :
                hour_index = d[date_index+8:date_index+10]
                if hour_index == '13':
                    shutil.copyfile(d, result_dir+d[FA_index:])
                    result_list.append(d[FA_index:])
                    count+=1
                    temp_module_index = module_index
                    print(count, d[FA_index:])
                
                
    print(len(result_list))
        
    
```
